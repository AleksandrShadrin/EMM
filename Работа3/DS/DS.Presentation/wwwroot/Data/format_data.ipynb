```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./Microsoft_Stock.csv")
df.head(10)
```

```python
df['Date'] = df['Date'].apply(lambda str : str.split(' ')[0].strip())
df['Date'] = pd.to_datetime(df['Date'])
df.head(10)
```

```python
serie = df.resample(rule='M', on='Date')['Close'].mean()
```

```python
serie.to_csv("new_data.csv")
```
